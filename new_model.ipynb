```python
import numpy as np

def get_important_features(pca_model, feature_names):
    """
    Extracts top `n_top` features per PCA component based on loadings.
    """
    components = pca_model.components_
    important_features = []
    
    # Loop through each component and get top features
    for component in components:
        top_indices = np.argsort(np.abs(component))[::-1]
        important_features.extend(feature_names[top_indices])
    
    # Remove duplicates and return
    return list(set(important_features))
```

```python

user_features_array = user_df_filtered.drop(columns=['user_id']).values.astype(np.float32)
item_features_array = item_df_filtered.drop(columns=['id']).values.astype(np.float32)


user_id_to_idx = {uid: idx for idx, uid in enumerate(user_df_filtered['user_id'])}
item_id_to_idx = {iid: idx for idx, iid in enumerate(item_df_filtered['id'])}


valid_ratings = ratings[
    ratings['user_id'].isin(user_id_to_idx) &
    ratings['book_id'].isin(item_id_to_idx)
]
train_ratings, test_ratings = train_test_split(valid_ratings, test_size=0.2, random_state=42)

batch_size = 1024
optimizer = tf.keras.optimizers.Adam()

for epoch in range(10):
    for batch in np.array_split(valid_ratings, len(valid_ratings) // batch_size):
        
        user_indices = batch['user_id'].map(user_id_to_idx).values
        item_indices = batch['book_id'].map(item_id_to_idx).values
        
        
        X_user = user_features_array[user_indices]
        X_item = item_features_array[item_indices]
        
       
        X_user_tensor = tf.convert_to_tensor(X_user, dtype=tf.float32)
        X_item_tensor = tf.convert_to_tensor(X_item, dtype=tf.float32)
        y_true = tf.convert_to_tensor(batch['rating'].values, dtype=tf.float32)
        
       
        with tf.GradientTape() as tape:
            user_emb = user_NN(X_user_tensor)
            item_emb = item_NN(X_item_tensor)
            y_pred = tf.reduce_sum(user_emb * item_emb, axis=1)
            loss = tf.keras.losses.MSE(y_true, y_pred)
        
        # Backpropagation
        grads = tape.gradient(loss, user_NN.trainable_variables + item_NN.trainable_variables)
        optimizer.apply_gradients(zip(grads, user_NN.trainable_variables + item_NN.trainable_variables))
        
    print(f"Epoch {epoch+1}, Loss: {loss.numpy():.4f}")
        
```

```python
def recommend(user_id, top_n=10):
    # Get user embedding
    user_idx = user_id_to_idx[user_id]
    user_emb = user_NN(user_features_array[[user_idx]])
    
    # Calculate similarity with all items
    item_embs = item_NN(item_features_array)
    similarities = tf.matmul(user_emb, item_embs, transpose_b=True)
    
    # Get top-N items
    top_items = np.argsort(-similarities.numpy())[0][:top_n]
    return [item_df_filtered.iloc[i]['id'] for i in top_items]

# Example: Recommend for user 314
print(recommend(314))
```

```python
def evaluate_model(test_data):
    # Get features
    user_indices = test_data['user_id'].map(user_id_to_idx).values
    item_indices = test_data['book_id'].map(item_id_to_idx).values
    X_user = user_features[user_indices]
    X_item = item_features[item_indices]
    
    # Predict
    user_embs = user_NN(X_user)
    item_embs = item_NN(X_item)
    y_pred = tf.reduce_sum(user_embs * item_embs, axis=1)
    y_true = test_data['rating'].values
    
    # Calculate metrics
    mse = tf.keras.losses.MSE(y_true, y_pred).numpy()
    mae = tf.keras.losses.MAE(y_true, y_pred).numpy()
    
    return mse, mae

test_mse, test_mae = evaluate_model(test_ratings)
print(f"Test MSE: {test_mse:.3f}, Test MAE: {test_mae:.3f}")
```
