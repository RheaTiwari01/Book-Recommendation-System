```python
book_tags = book_tags.merge(tags, left_on='tag_id', right_on='tag_id')

# Merge the book_tags with ratings to associate ratings with genres
book_ratings = pd.merge(book_tags, ratings, left_on='goodreads_book_id', right_on='book_id')
```

```python
# Group by genre (tag_name) and calculate the average rating for each genre
genre_ratings = book_ratings.groupby('tag_name').agg(
    average_rating=('rating', 'mean'),  # Calculate the average rating per genre
    count=('rating', 'count')  # Count the number of ratings per genre
).reset_index()
genre_ratings
```

```python
top_200_genres = genre_ratings.sort_values('count', ascending=False).head(200)

# Display the top 200 genres with their average rating
print(top_200_genres)
```

```python
# 1. Find the top 200 genres by count
top_200_genres = book_ratings['tag_name'].value_counts().nlargest(200).index

# 2. Filter the book_ratings DataFrame to only include those top 200 genres
filtered_ratings = book_ratings[book_ratings['tag_name'].isin(top_200_genres)]

# 3. Calculate the overall average rating per user
user_avg_rating = filtered_ratings.groupby('user_id').agg(
    total_avg_rating=('rating', 'mean')  # User's total average rating
).reset_index()

# 4. Calculate the average rating per user for each genre
user_genre_avg = filtered_ratings.groupby(['user_id', 'tag_name']).agg(
    genre_avg_rating=('rating', 'mean')  # User's average rating for each genre
).reset_index()

# 5. Pivot the table to have a column for each genre's average rating
user_genre_pivot = user_genre_avg.pivot(
    index='user_id',  # Rows represent each user
    columns='tag_name',  # Each genre becomes a separate column
    values='genre_avg_rating'  # Values are the genre average ratings
).reset_index()

# 6. Merge the overall average rating with the genre-wise ratings
user_ratings = pd.merge(user_avg_rating, user_genre_pivot, on='user_id')

# Final result: user_ratings DataFrame with total average rating and genre-wise average ratings
print(user_ratings.head())  # Show the first few rows
user_ratings
```

```python
import numpy as np
import pandas as pd
import tensorflow as tf
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler

# Example rating DataFrame (must have the same length as user_df and y)
# Create a DataFrame from your ratings data
ratings = pd.DataFrame({
    'user_id': user_df['user_id'],            # Assume this corresponds to ratings
    'item_id': item_df['goodreads_book_id'],  # Link this correctly to the ratings
    'rating': y                                 # Your ratings Series
})

# Ensure ratings DataFrame is aligned
# Drop any rows with missing values to ensure consistency
ratings.dropna(inplace=True)

# Split the data into train and test sets
user_train, user_test, item_train, item_test, y_train, y_test = train_test_split(
    ratings['user_id'], ratings['item_id'], ratings['rating'], 
    train_size=0.80, shuffle=True, random_state=1
)

# Normalize ratings
y_train_np = y_train.values.reshape(-1, 1)
y_test_np = y_test.values.reshape(-1, 1)

# Fit and transform the reshaped array for normalization
scaler = MinMaxScaler(feature_range=(-1, 1))
ynorm_train = scaler.fit_transform(y_train_np)
ynorm_test = scaler.transform(y_test_np)

# Number of unique users and items
num_users = ratings['user_id'].nunique()  # Get unique user count
num_items = ratings['item_id'].nunique()  # Get unique item count

# Embedding dimension
embedding_dim = 32

# Building embedding layers for users and items
user_input = tf.keras.layers.Input(shape=(1,), name='user_input')  # Single index input for users
item_input = tf.keras.layers.Input(shape=(1,), name='item_input')  # Single index input for items

# Embedding layers
user_embedding = tf.keras.layers.Embedding(input_dim=num_users, output_dim=embedding_dim)(user_input)
item_embedding = tf.keras.layers.Embedding(input_dim=num_items, output_dim=embedding_dim)(item_input)

# Flatten the embeddings
user_vecs = tf.keras.layers.Flatten()(user_embedding)
item_vecs = tf.keras.layers.Flatten()(item_embedding)

# Optionally, you can add dense layers after the embeddings
user_vecs = tf.keras.layers.Dense(64, activation='relu')(user_vecs)
item_vecs = tf.keras.layers.Dense(64, activation='relu')(item_vecs)

# Normalize the vectors
user_vecs = tf.linalg.l2_normalize(user_vecs, axis=1)
item_vecs = tf.linalg.l2_normalize(item_vecs, axis=1)

# Dot product for similarity score
output = tf.keras.layers.Dot(axes=1)([user_vecs, item_vecs])

# Define and compile the model
model = tf.keras.Model(inputs=[user_input, item_input], outputs=output)
model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=0.01),
              loss=tf.keras.losses.MeanSquaredError())

# Training the model
model.fit([user_train, item_train], ynorm_train, 
          validation_data=([user_test, item_test], ynorm_test), 
          epochs=50, 
          batch_size=32)
```
